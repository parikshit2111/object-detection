```python
import cv2
import dlib
```

```python
from google.colab.patches import cv2_imshow
input=cv2.imread('drive/MyDrive/Face detection/task1/people1.jpg')
input = cv2.resize(input, (input.shape[1] // 2, input.shape[0] // 2))
cv2_imshow(input)
```

```python

gray_image=cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
gray_image
cv2_imshow(gray_image)
```

```python
body_model=cv2.CascadeClassifier('drive/MyDrive/Face detection/task1/haarcascade_frontalface_default.xml')
```

```python
bodys=body_model.detectMultiScale(gray_image,scaleFactor=1.5, minNeighbors=3)
```

```python
faces=bodys
faces
```

```python
import numpy as np
for (i, rect) in enumerate(bodys):
    x,y,w,h=rect
    #draw a rectangle
    cv2.rectangle(input, (x, y), (x + w, y + h), (0, 0, 255), 2)
```
