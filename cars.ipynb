```python
import cv2
import dlib
```

```python
input=cv2.imread('drive/MyDrive/Face detection/task1/car.jpg')
input
```

```python
#grey

gray=cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
gray
```

```python
car_model=cv2.CascadeClassifier('drive/MyDrive/Face detection/task1/cars.xml')
```

```python
cars=car_model.detectMultiScale(gray,scaleFactor=1.05, minNeighbors=3)
```

```python
cars
```

```python
import numpy as np
for (i, rect) in enumerate(cars):
    x,y,w,h=rect
    #draw a rectangle
    cv2.rectangle(input, (x, y), (x + w, y + h), (0, 255, 0), 2)
```
