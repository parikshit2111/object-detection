```python
import cv2
import dlib
```

```python
input=cv2.imread('drive/MyDrive/Face detection/task1/clock.jpg')
input
```

```python

gray_image=cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
gray_image
```

```python
clock_model=cv2.CascadeClassifier('drive/MyDrive/Face detection/task1/clocks.xml')
```

```python
clocks=clock_model.detectMultiScale(gray_image,scaleFactor=1.005, minNeighbors=3)
```

```python
clocks
```

```python
import numpy as np
for (i, rect) in enumerate(clocks):
    x,y,w,h=rect
    #draw a rectangle
    cv2.rectangle(input, (x, y), (x + w, y + h), (0, 0, 255), 2)
```
